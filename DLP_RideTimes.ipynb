```python
# Imports
import time
import random
import json
import requests
import pytz
import datetime

#Urls
url = "https://queue-times.com/parks.json"
dlpQurl = "https://queue-times.com/parks/4/queue_times.json"
wdspQurl = "https://queue-times.com/parks/28/queue_times.json"

paris = pytz.timezone('Europe/Paris')
parisDatetime = datetime.datetime.now(paris)
print (parisDatetime.strftime("%b %d %H:%M"))
```

```python
# This is the Park details load

r = requests.get(url)
#print(r.json())

parks = r.json()

""" for i in range(0,len(parks)):
    print(i)
    print(parks[i])
    print(len(parks[i]["parks"])) """

for p in range(0,len(parks[14]["parks"])):
    if parks[14]["parks"][p]["country"] == "France":# and parks[12]["parks"][p]["timezone"] == "America/New_York":
        print(parks[14]["parks"][p])
        print("---------------------------------------------------------------")
#        time.sleep(0.3)
```

```python

### Walt Disney Studios Paris

wdspR = requests.get(wdspQurl)

wdsp = wdspR.json()

#print(wdsp['lands'])

for a in range(0,len(wdsp['lands'])):
  print(wdsp['lands'][a]['name'],':')
  for r in range(0,len(wdsp['lands'][a]['rides'])):
    print(' ',wdsp['lands'][a]['rides'][r]['name'], '\n wait time: ' , wdsp['lands'][a]['rides'][r]['wait_time'], '- is open:' , wdsp['lands'][a]['rides'][r]['is_open'], '\n---------------')
```
